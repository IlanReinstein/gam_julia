### Cubic Splines Example

```julia
function Rk(x, z)
    a = (((z' .- 0.5).^2 .- 1/12) .* ((x .- 0.5).^2 .- 1/12)) ./ 4
    b = ((abs.(x .- z') .- 0.5).^4 - 0.5 .* (abs.(x .- z') .- 0.5).^2 .+ 7/240)./24
    return a .- b
end
```

```julia
function spl_X(x, xk)
    q = length(xk) + 2
    n = length(x)
    X = ones(n, q)
    X[:,2] = x
    X[:,3:q] = Rk(x, xk)
    return X
end
```

```julia
xk = 1:4
X = spl_X(x, xk/5)
mod1 = fit(LinearModel, X, wear);
```

```julia
xp = 0:0.01:1
Xp = spl_X(xp, xk/5)
scatter(x, wear);
plot!(xp, Xp * coef(mod1))
```

**Notes:** 
- The degree of smoothness, controlled by `q = (n_knots + 2)` is arbitrary
- The model fitted to the basis expansion is still a linear model

### Penalized Regression Spline

```julia
function spl_S(xk)
    q = length(xk) + 2
    S = zeros(q, q)
    S[3:q, 3:q] = Rk(xk, xk)
    return S
end
```

```julia
# Optional Function - No need to decompose S to apply sqrt
function mat_sqrt(S)
    evals, evec = eigen(Symmetric(S))
    return evec * Diagonal(evals) .^ 0.5 * evec'
end
```

```julia
# Fit Penalized Regression Spline
function prs_fit(y, x, xk, lambda)
    q = length(xk) + 2
    n = length(x)
    Xa = [spl_X(x, xk); sqrt(spl_S(xk)) * sqrt(lambda)]
    y = [y; zeros(q)]
    return lm(Xa, y)
end
```

```julia
lambda = 0.00001
xk = 1:7
mod2 = prs_fit(wear, x, xk/8, lambda)
Xp = spl_X(xp, xk/8)
scatter(x, wear);
plot!(xp, Xp * coef(mod2))

```

### Using Cross-Validation to determine the best $\lambda$

```julia
lambda = 1e-8
n = length(wear)
V = zeros(60);
```

```julia
for i = 1:60
    m = prs_fit(wear, x, xk/8, lambda)
    Xm = m.pp.X
    A = Xm * inv(Xm' * Xm) * Xm'
    trA = sum(diag(A)[1:n])
    rss = sum((wear - m.rr.mu[1:n]).^2)
    V[i] = n*rss/(n - trA)^2
    lambda = lambda * 1.5
end
```

```julia
l = @layout [a b]
p1 = plot(1:60, V, title = "GCV Score")
i = argmin(V)
mod3 = prs_fit(wear, x, xk/8, 1.5 ^ (i - 1)*1e-8)
Xp = spl_X(xp, xk/8)
p2 = scatter(x, wear, title = "GCV Optimal Fit")
p2 = plot!(xp, Xp * coef(mod3))
plot(p1, p2, layout = l)
```

### 3.3 Additive Models

```julia
using RDatasets
using Statistics
```

```julia
function am_setup(x, z, q = 10)
    xk = quantile(unique(x), collect(1:(q-2))/(q-1))
    zk = quantile(unique(z), collect(1:(q-2))/(q-1))
    S = zeros(2, 2q - 1, 2q - 1)
    S[1, 2:q, 2:q] = spl_S(xk)[Not(1), Not(1)]
    S[2,(q + 1):(2q - 1), (q+1):(2q - 1)] = spl_S(zk)[Not(1), Not(1)]
    n = length(x)
    X = ones(n, 2q - 1)
    X[:, 2:q] = spl_X(x, xk)[:, Not(1)]
    X[:, (q + 1):(2q - 1)] = spl_X(x, xk)[:, Not(1)]
    return X, S
end
```

```julia
function minmax(x)
    min = minimum(x)
    max = maximum(x)
    return (x .- min)/(max .- min)
end
```

```julia
function fit_am(y, X, S, sp)
    rS = sqrt(sp[1]*S[1, :, :] + sp[2]*S[2, :, :])
    n, q = size(X)
    X1 = vcat(X, rS)
    y1 = y
    y1 = [y1; zeros(q)]
    # X1 is rank-deficient
    # lm Does not work here
    beta = X1\y1
    mu = X1*beta
#     Xm = X1[1:n,:]
    A = X * inv(X' * X) * X'
    trA = sum(diag(A))
    norm = sum((y - mu[1:n]).^2)
    return Dict("model" => beta, "gcv" => (norm*n)/((n - trA)^2), "sp" => sp)
end
```

```julia
trees = dataset("datasets", "trees")
trees.Girth = minmax(trees.Girth)
trees.Height = minmax(trees.Height)
x = trees.Girth
z = trees.Height
y = trees.Volume
X, S = am_setup(x, z);
```

```julia
sp = zeros(2)
s0 = 1e-5
best = zeros(30, 30);
```

```julia
for i = 1:30, j = 1:30
    sp[1] = s0 * 2 ^ (i - 1)
    sp[2] = s0 * 2 ^ (j - 1)
    b = fit_am(trees.Volume, X, S, sp)
    best[i, j] = b["gcv"]
#     if i + j == 2 
#         best = b
#     elseif b["gcv"] < best["gcv"]
#         best = b
#     end
end
```

## 3.4 Generalized Additive Models

```julia
function fit_gam(y, X, S, sp)
    rS = sqrt(sp[1]*S[1, :, :] + sp[2]*S[2, :, :])
    n, q = size(X)
    X1 = vcat(X, rS)
    b = zeros(q)
    b[1] = 1
    norm = 0
    old_norm = 1

    while abs(norm - old_norm) > 1e-4*norm
        eta = (X1*b)[1:n]
        mu = exp.(eta)
        z = (y - mu)./mu + eta
        z1 = [z; zeros(q)]
        b = X1\z1
        A = X * inv(X'* X) * X'
        trA = tr(A)
        y_hat = X1 * b
        norm = sum((z1 - y_hat)[1:n].^2)
    end
    return Dict("model" => b, "gcv" => (norm*n)/((n - trA)^2), "sp" => sp)
end
```

```julia
sp = zeros(2)
s0 = 1e-5
best = zeros(30, 30);
for i = 1:30, j = 1:30
    sp[1] = s0 * 2 ^ (i - 1)
    sp[2] = s0 * 2 ^ (j - 1)
    b = fit_gam(trees.Volume, X, S, sp)
    best[i, j] = b["gcv"]
#     if i + j == 2 
#         best = b
#     elseif b["gcv"] < best["gcv"]
#         best = b
#     end
end
```
